```python
# cctv 데이터 가져오기

cctv_data = pd.read_csv('C:/Users/hmkk0/.ipynb_checkpoints./서울특별시 (안심이) CCTV 설치 현황.csv', encoding = 'CP949')
cctv_data
```

```python
#'CCTV 용도'로 유일한 값 찾기

cctv_data['CCTV 용도'].unique()
```

```python
#데이터 결측값 채우기

cctv_data.fillna({'CCTV 용도':'고정식'}, inplace=True)
```

```python
# cctv 용도에 맞추어 값 가져오기

condition1 = cctv_data['CCTV 용도'] == '방범'
condition2 = cctv_data['CCTV 용도'] == '고정식'
condition3 = cctv_data['CCTV 용도'] == '회전식'
condition4 = cctv_data['CCTV 용도'] == '어린이안전'

cctv_filter = cctv_data[condition1 | condition2 |  condition3 | condition4]

cctv_filter.head(10)
```

```python
#'자치구', 'CCTV 수량'만 필터링 하기

cctv= cctv_filter[['자치구','CCTV 수량']]

cctv
```

```python
#자치구별 'CCTV 수량' 합치기

cctv_sum= cctv.groupby(cctv_filter['자치구']).sum()

cctv_sum
```

```python
## 지역별 인구밀도

people = pd.read_excel('C:/Users/hmkk0/.ipynb_checkpoints./서울 인구.xls', usecols = 'B,E', skiprows =[1])
                      
people = people.set_index('지역')
    
people
```

```python
## 지역별 인구밀도 가나다 순 sorting

people = people.sort_index()

people
```

```python
# 인구밀도 CCTV 합치기

result = pd.concat([people,cctv_sum['CCTV 수량']],axis=1)

result
```
